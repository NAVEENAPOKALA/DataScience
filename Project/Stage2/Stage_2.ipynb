#### **a. Calcuate the selected features mean, median, variance, skewness and kurtosis. (10 points)**

**Features I will be performing mean, median, variance, skewness and kurtosis include views_day1,likes_day1, dislikes_day1, comments_day1, likes_ratio, comment_ratio.**

```python
def task_1 (df):
    columns_1=['views_day1','likes_day1', 'dislikes_day1', 'comments_day1', 'likes_ratio', 'comments_ratio']
    mean=df[columns_1].mean()
    median=df[columns_1].median()
    variance=df[columns_1].var()
    skw=df[columns_1].skew()
    krt=df[columns_1].kurtosis()

    final_result= pd.DataFrame ({'mean': mean, 'median': median,'variance': variance, 'skew': skw,'kurtosis': krt})
    return final_result
```

### Task 2 - Distribution Fitting (20 points) --- Xiuwen

#### **a. Select one suitable numerical feature from task 1.  Propose a theoretical distribution (Normal, Binomial, Poisson, Exponential, Geometric, etc.) and explain why you choose it. (6 points)**

Selected Feature: Views Day 1

Proposed Distribution: Gamma Distribution

Reasoning:
1. Views data is strictly positive (Gamma supports positive values only)
2. The data is right-skewed with a heavy tail (Gamma can model this)
3. Real-world view counts often follow power-law or Gamma-like distributions
4. Gamma distribution is flexible with shape and scale parameters

```python
# Select 'views_day1' for distribution fitting
views_data = df['views_day1']
```

#### **b. Plot histogram of your feature, overlay a smooth curve of the fitted theoretical distribution (10 points). Discuss how your data fits the distribution. (4 points)**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm, poisson, expon, gamma
import warnings
warnings.filterwarnings('ignore')

# Fit Gamma distribution to the data
params = gamma.fit(views_data)
x = np.linspace(views_data.min(), views_data.max(), 1000)
pdf_fitted = gamma.pdf(x, *params)

# Plot
plt.figure(figsize=(12, 6))

# Histogram
plt.hist(views_data, bins=50, density=True, alpha=0.7, color='skyblue', label='Data Histogram')

# Fitted distribution
plt.plot(x, pdf_fitted, 'r-', linewidth=2, label=f'Fitted Gamma Distribution')

plt.xlabel('Views Day 1', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Task 2b - Distribution Fitting for Views Day 1\nGamma Distribution Fit', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Goodness of fit assessment
print("\nGoodness of Fit Assessment:")
ks_stat, ks_pvalue = stats.kstest(views_data, 'gamma', args=params)
print(f"Kolmogorov-Smirnov Test: statistic = {ks_stat:.4f}, p-value = {ks_pvalue:.4f}")

# Calculate mean and variance of fitted distribution
shape, loc, scale = params
fitted_mean = shape * scale + loc
fitted_var = shape * scale**2

print(f"\nData Mean: {views_data.mean():.2f}, Fitted Gamma Mean: {fitted_mean:.2f}")
print(f"Data Variance: {views_data.var():.2f}, Fitted Gamma Variance: {fitted_var:.2f}")
```

### Task 3 - Point Estimate (20 points) --- Xiuwen

#### **a. Randomly select 100 samples from the feature selected in Task 2, caculate the sample mean and variance. Compare with the population mean and variance. (10 points)**

```python
# Set seed for reproducibility
np.random.seed(42)

# Random sample of 100
sample_100 = np.random.choice(views_data, size=100, replace=False)

# Calculate sample statistics
sample_mean = np.mean(sample_100)
sample_variance = np.var(sample_100, ddof=1)  # Sample variance

# Population statistics
population_mean = views_data.mean()
population_variance = views_data.var()

print("Task 3a - Sample vs Population Comparison:")
print("=" * 50)
print(f"Sample Size: 100")
print(f"Sample Mean: {sample_mean:,.2f}")
print(f"Population Mean: {population_mean:,.2f}")
print(f"Difference: {abs(sample_mean - population_mean):,.2f} ({abs((sample_mean - population_mean)/population_mean)*100:.2f}%)")
print(f"\nSample Variance: {sample_variance:,.2f}")
print(f"Population Variance: {population_variance:,.2f}")
print(f"Difference: {abs(sample_variance - population_variance):,.2f} ({abs((sample_variance - population_variance)/population_variance)*100:.2f}%)")
```

#### **b. Repeat the random sampling 3 times and show how sample mean/variance vary due to randomness. Discuss your findings. (10 points)**

```python
print("\nTask 3b - Multiple Sampling Results:")
print("=" * 50)

sampling_results = []

for i in range(3):
    np.random.seed(i*10)  # Different seeds for different samples
    sample = np.random.choice(views_data, size=100, replace=False)

    sample_mean = np.mean(sample)
    sample_var = np.var(sample, ddof=1)

    sampling_results.append({
        'Sample': i+1,
        'Mean': sample_mean,
        'Variance': sample_var,
        'Mean_Diff_Pct': abs((sample_mean - population_mean) / population_mean * 100),
        'Var_Diff_Pct': abs((sample_var - population_variance) / population_variance * 100)
    })

    print(f"\nSample {i+1}:")
    print(f"  Mean: {sample_mean:,.2f} (Difference: {abs(sample_mean - population_mean):,.2f}, {abs((sample_mean - population_mean)/population_mean)*100:.2f}%)")
    print(f"  Variance: {sample_var:,.2f} (Difference: {abs(sample_var - population_variance):,.2f}, {abs((sample_var - population_variance)/population_variance)*100:.2f}%)")

# Create summary table
results_df = pd.DataFrame(sampling_results)
print("\n" + "="*50)
print("Summary of Sampling Variation:")
print(results_df.round(2))
```

 Sample Mean Variability: The sample means range from 1,021,582 to 1,055,952, with differences from the population mean ranging from 1.64% to 4.84%. This shows reasonable accuracy in estimating the population mean with samples of size 100.

Sample Variance Variability: The sample variances show much greater variability (23-38% differences from population variance). This is expected because: Variance is more sensitive to outliers than the mean. The highly skewed distribution with extreme outliers makes variance estimation challenging. Small samples may miss or over-represent extreme values.

Sampling Implications:
For highly skewed distributions like video views, larger sample sizes would be needed for stable variance estimates. The mean is relatively more stable across samples compared to variance. The Law of Large Numbers suggests that with larger samples, these estimates would converge to the population parameters.
