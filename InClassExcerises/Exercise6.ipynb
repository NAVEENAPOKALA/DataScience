## Broadcasting operations

```python
## converting the data into a numpy array
arr1 = np.array(data)
print(arr1)
```

```python
## Adding a constant value to the arr1
arr1 = arr1+5
print(arr1)
```

```python
# subtracting a vector [1,2,3,4] from each row of the matrix
arr2 = np.array([1,2,3,4])
subarr = arr1-arr2
print(subarr)
```

```python
# multiply a column vector [1,2,3,4] from each column of the matrix
arr3 = np.array([[1],[2],[3],[4]])
print(arr3)
mularr = arr1*arr3
print(mularr)
```

```python
# subtracting the mean of each row from the matrix
meanarr = np.mean(arr1,axis=1)
print(meanarr)
```

```python
print(arr1)
submean = arr1-meanarr.reshape(-1, 1)
print(submean)
```
