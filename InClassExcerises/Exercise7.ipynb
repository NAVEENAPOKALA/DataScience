```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
n=100
p=1/6
size=1000
samples = binom.rvs(n=n,p=p,size=size,random_state=42)
#plt.hist(samples,bins=20,edgecolor='black')
plt.hist(samples,bins=20,density=True,edgecolor='black',alpha=0.7)
plt.show()
```
