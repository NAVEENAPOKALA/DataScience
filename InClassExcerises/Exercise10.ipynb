```python
import numpy as np
from scipy import stats
```

```python
sample_crimerate = [100,23,34,5,67,89,34,33,20]
```

```python
t_statistic,p_value = stats.ttest_1samp(sample_crimerate,30)
```

```python
print("t statistic " , t_statistic)
print("p_value " , p_value)
```

```python
old_cholesterol = [54,,23,34,32,45,55,34,33,20,33]
```

```python
new_cholesterol = [45,33,22,10,8,7,6,5,41,22]
```

```python
t_statistic,p_value = stats.ttest_ind(new_cholesterol,old_cholesterol)
```

```python
print("t statistic " , t_statistic)
print("p_value " , p_value)
```
