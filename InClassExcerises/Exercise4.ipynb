```python
import pandas as pd
import numpy as np

np.random.seed(24)

data = pd.Series(np.random.normal(loc=50,scale=5,size=50))

outliers = pd.Series([100,110,120,130])

data_with_outliers = pd.concat([data,outliers], axis = 0)

df = pd.DataFrame(data_with_outliers,columns=['values'])

print(df.describe())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['normalized_values'] = scaler.fit_transform(df[['values']])
print(df)
```

```python
df['Z-score'] = (df['values'] - df['values'].mean()) /df['values'].std()
print(df['Z-score'])
```

```python
df['log_values'] = np.log1p(df['Z-score'])
print(df)
```

```python
import pandas as pd
import numpy as np

np.random.seed(24)

data = pd.Series(np.random.normal(loc=50,scale=5,size=50))

outliers = pd.Series([100,110,120,130])

data_with_outliers = pd.concat([data,outliers], axis = 0)

df1 = pd.DataFrame(data_with_outliers,columns=['values'])

print(df1.describe())
```

```python
scaler = MinMaxScaler()
df1['normalized_values'] = scaler.fit_transform(df1[['values']])
print(df1)
```

```python
df1['log_values'] = np.log1p(df1['values'])
print(df1)
```

```python
df1['Z-score'] = (df1['values'] - df1['values'].mean()) /df1['values'].std()
print(df1['Z-score'])
```

```python
df1.loc[(df1['Z-score']<3) & (df1['Z-score'] > -3)]
```
