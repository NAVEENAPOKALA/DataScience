```python
def factorial(n):
    if n == 0 :
        return 1
    else :
        return n*factorial(n-1)
```

```python
factorial(5)
```

## Factorial Function Explanation
Fcatorial of a number is the **product of all the numbers starting from 1 to the given number**. To write a factorial function, we are using a **recursive** approach. Recursive function is a function that calls itself. With each recursive call, we are reducing the number by 1 and this repeats it reaches one.
